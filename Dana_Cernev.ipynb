```python
url = "https://github.com/arnautoleg/Dana_Cernev/raw/main/teza%20master%20statistica%20(2)%20(1).xls"
df=pd.read_excel(url, sheet_name=2)

df['incidence'] = round(df['cazuri noi']/df['copii pinala 15 ani']*100000, 2)

df
```

```python
df['prevalence'] = round(df['nr.adresari'][df['nr.adresari']>0]/df['copii pinala 15 ani'] * 100000, 1)

df
#= round(df['cazuri noi']]*100000, 2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['<1 per year', '1 per year', '2 per year', '3 per year', '>3 per year', 'Total']
b = [8, 41, 25, 14, 10, 98]
f = [7, 28, 36, 15, 14, 100]
t = [15, 69, 61, 29, 24, 198]

# Set the bar width and positions
bar_width = 0.2
index = np.arange(len(categories))

# Create the grouped bar plot
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.bar(index - bar_width, b, bar_width, label='M', color='b', align='center')
plt.bar(index, f, bar_width, label='F', color='g', align='center')
plt.bar(index + bar_width, t, bar_width, label='Total', color='r', align='center')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Recidives, absolute values')
plt.xticks(index, categories)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
